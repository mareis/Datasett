### Underveisoppgave 5.4.1



*  Importer pandasbiblioteket

*  lag variabelen **bil** for de [loggede dataene for strekning og tid](https://raw.githubusercontent.com/mareis/datafiler/main/bil_4_5_2.csv). 


*  Skriv ut dataene til de øverste radene, slik at du ser strukturen til dataene. 

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/mareis/Datasett/main/Egne/bil_4_5_skraplan.csv', sep=';')
df.head()
```

```python
tid = df['tid']
strekning = df ['strekning']
```

```python
# Løs oppgaven her
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

plt.figure(figsize=(8,4))
plt.plot(tid, strekning, '.')

plt.title("Strekningen til bilen")
plt.xlabel('Tid s')
plt.ylabel('Strekning (m)')
plt.show()
```

```python

import numpy as np

modell = np.polyfit(tid, strekning, 2)
modell
```

```python
a, b, c = modell
```

```python
# Løs oppgaven her
def s(t):
    return a*t**2 + b*t + c

plt.figure(figsize=(8,4))
plt.plot(tid, strekning, '.', label="Sensordata")
plt.plot(tid, s(tid), label="$s(t)$")

plt.title("Strekningen til bilen")
plt.xlabel('Tid (s)')
plt.ylabel('Strekning (m)')
plt.legend()
plt.show()
```

```python
def derivert(f, x, h = 10**-6):
    return (f(x + h) - f(x))/h

s_derivert = derivert(s, tid)

plt.figure(figsize=(8,4))
plt.plot(tid, s_derivert)

plt.title("Hastigheten til bilen")
plt.xlabel('tid s')
plt.ylabel('hastighet m/s')
plt.show()
```
