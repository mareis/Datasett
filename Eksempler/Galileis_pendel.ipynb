Den eksperimentelle fysikkens far, italieneren Galileo Galilei (1564–1642), fant gjennom eksperimenter at en pendels svingetid bare er bestemt av pendelens lengde.



```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/mareis/Datasett/main/Egne/pendel.csv', sep=';')
df
```

```python
x = df['lengde']
y = df['tid']
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8,4))
plt.plot(x, y, '.')
plt.title("Pendelforsøk")
plt.xlabel('Lengden til pendelen (m)')
plt.ylabel('Tid (s)')
plt.show()
```

Målenøyaktigheten kan variere en del som vi ser over. Men det er en klar sammenheng mellom lengden på pendelen og svingtiden.<br><br>


## En modell

Det er en matematisk formel for å regne ut en pendels svingetid, når en ser bort fra all friksjon:

$$T(l) = 2\pi\sqrt{\frac{l}{g}}, \text{ der } g = 9.81 \text{ m/s}$$

<br>

For å se om det stemmer så kan vi plotte grafen sammen med punktene.

```python
import numpy as np

def T(l):
    return 2*np.pi*np.sqrt(l/9.81)

l = np.linspace(0.1, 1, 100)


plt.figure(figsize=(8,4))
plt.plot(x, y, '.', label="Sensordata")
plt.plot(l, T(l), label="$T(l)$")

plt.title("Pendel")
plt.xlabel('Lengde til pendelen (m)')
plt.ylabel('Tid (s)')
plt.legend()
plt.show()
```

## Den deriverte

Vi kan nå bruk Newtons kvotient til å bestemme $T'(l)$ og vi plotter resulatet.


```python
def derivert(f, x, h = 10**-6):
    return (f(x + h) - f(x))/h

T_derivert = derivert(T, l)

plt.figure(figsize=(8,4))
plt.plot(l, T_derivert)

plt.title("Den deriverte")
plt.xlabel('Lengde til pendel (m)')
plt.ylabel('Tid (s/m)')
plt.show()
```
